```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import pickle
```

```python
df = pd.read_excel('../data/raw/Фотокатализ.xlsx')
df
```

```python
df.drop(df.columns[0], axis=1, inplace=True)
df
```

```python
le_prec = LabelEncoder()
df['Прекурсор '] = le_prec.fit_transform(df['Прекурсор '])
with open('../models/preprocessing/le_prec.pkl', 'wb') as f:
    pickle.dump(le_prec, f)
    f.close()
df
```

```python
le_dob = LabelEncoder()
df['Добавка'] = le_dob.fit_transform(df['Добавка'])
with open('../models/preprocessing/le_dob.pkl', 'wb') as f:
    pickle.dump(le_dob, f)
    f.close()
df
```

```python
le_method = LabelEncoder()
df['Метод нанесения pt'] = le_method.fit_transform(df['Метод нанесения pt'])
with open('../models/preprocessing/le_method.pkl', 'wb') as f:
    pickle.dump(le_method, f)
    f.close()
df
```

```python
le_donor = LabelEncoder()
df['Донор электронов'] = le_donor.fit_transform(df['Донор электронов'])
with open('../models/preprocessing/le_donor.pkl', 'wb') as f:
    pickle.dump(le_donor, f)
    f.close()
df
```

```python
df.to_csv('../data/processed/data.csv', index=False)
```
