```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('../data/trained/grid_search_results.json', 'r') as f:
    grid_search_results = json.loads(f.read())
    f.close()
    
grid_search_results = pd.json_normalize(grid_search_results)
grid_search_results
```

# Train RMSE

```python
plt.plot(grid_search_results['cv_results.iterations'].values[0], grid_search_results['cv_results.train-RMSE-mean'].values[0], label='Mean RMSE')
# plt.plot(grid_search_results['cv_results.iterations'].values[0], grid_search_results['cv_results.train-RMSE-std'].values[0], label='Std RMSE')
plt.title('Mean train RMSE')
plt.xlabel('Iterations')
plt.ylabel('RMSE')
plt.legend()
plt.tight_layout()
plt.savefig('../reports/figures/mean_train_rmse.svg', format='svg', dpi=1200)
```

# Test RMSE

```python
plt.plot(grid_search_results['cv_results.iterations'].values[0], grid_search_results['cv_results.test-RMSE-mean'].values[0], label='Mean RMSE')
# plt.plot(grid_search_results['cv_results.iterations'].values[0], grid_search_results['cv_results.train-RMSE-std'].values[0], label='Std RMSE')
plt.title('Mean test RMSE')
plt.xlabel('Iterations')
plt.ylabel('RMSE')
plt.legend()
plt.tight_layout()
plt.savefig('../reports/figures/mean_test_rmse.svg', format='svg', dpi=1200)
```
